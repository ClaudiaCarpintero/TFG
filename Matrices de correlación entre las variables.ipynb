```python
from mlxtend.plotting import heatmap
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

Matriz de correlación de las variables mecánicas


```python
dataSet = list()
ciclistas = open ('/Users/34603/Desktop/TFG/Python/Archivo completo.txt','r')
j=0
for linea in ciclistas: #Recorremos cada linea del archivo
    caracteristicas = linea.split()
    if j==0:
        nombresVariables=caracteristicas[2:] 
    else:
        dataSet.append([float(x) for x in caracteristicas[2:]])
    j+=1
data = np.array(dataSet)
df = pd.DataFrame(dataSet, columns = nombresVariables)
print(df.shape)
#Solo cojo algunas variables para ver si están correlacionadas:
df_small = df.iloc[:,[3,4,5,6,7,18,24,28,34,39,40,41]]
correlation_mat = df_small.corr()
sns.heatmap(correlation_mat, annot = True)
plt.show()
```
